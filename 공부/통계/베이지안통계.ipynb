<h1><b>불량품 추정문제</b></h1>

<h3>어느 공장에서 제품을 생산하는데, 기계의 상태가 양호하면 불량률 7%, 그렇지 않으면 불량률 30%이며,<br><br>
평균적으로 기계가 양호한 경우 90%이다. 기계의 상태가 양호하지 않으면 제품생산을 멈추고 수리를 하는데,<br><br>
어느 날 제품생산 결과가 다음과 같다. 기계의 상태에 대한 사후확률을 구하시오.</h3>

<h4>제품 생산결과 : [G, G, B, G, B, G, G, G, G : G-Good, B-Bad]</h4>

<h4>문제의 목적 : 양품, 불량품에 따라 기계가 양호한지 알고싶다</h4>

<h4>- P(기계상태 양호|양품)</h4>
<h4>- P(기계상태 불량|불량)</h4>

H1 : 기계 양호
H2 : 기계 불량

사전분포prior 기계양호 0.9/기계불량0.1
우도는 제품 생산결과에 따라 다름

```r
# 기계 상태 양호 likelihood
G_likelihood = 0.93 #P(양품|양호)
B_likelihood = 0.07 #P(불량품|양호) 1-0.93
```

```r
baysian_update = function(state, prior){
    if(state == 1){
        posterior = (G_likelihood * prior)/((G_likelihood * prior) + (0.7 * (1-prior))) # P(양호|양품)
    }else{
        posterior = (B_likelihood * prior)/((B_likelihood * prior) + (0.3 * (1-prior))) # P(양호|불량품)
    }
    ret = posterior
    return(ret)
}
```

```r
# 1: 양품 2: 불량품
new_data = c(1,1,0,1,0,1,1,1,1)
```

```r
# updated probability 초기값
updated = rep(0, length(new_data)+1)

# prior
updated[1] = 0.9
```

```r
# bayesian update
for (i in 1:length(new_data)){
    updated[i+1] = baysian_update(new_data[i],updated[i])
}
```

```r
# table만듦
update_count = 1:length(updated)-1
update_count = paste(update_count,'회',sep='')
update_mat = matrix(updated, 1, length(updated))
update_mat = round(update_mat,3)
update_mat = rbind(update_mat,(1-update_mat))
colnames(update_mat) = update_count
row.names(update_mat) = c('H1','H2')
```
