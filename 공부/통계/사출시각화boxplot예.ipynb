```python
import pandas as pd
import matplotlib.pyplot as plt
file_path = 'C:/injection.csv'
df1 = pd.read_csv(file_path, encoding='CP949')
print(df1)
```

```python
df1.head()
```

```python
peak=df1['PEAK압력']
plt.plot(peak.index,peak.values)
```

```python
from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/malgun.ttf"
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font',family=font_name)
plt.figure(figsize=(10,5))
plt.xticks(rotation='vertical')
plt.plot(peak)

plt.title('PEAK압력')
plt.xlabel('시간')
plt.ylabel('압력')

plt.legend(labels=['전체'], loc='best')
plt.show()
```

```python
#화면 분할 그래프
cu=df1['최소쿠션']

plt.style.use('ggplot')
fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)

ax1.plot(peak, marker='o', markersize = 10, label = 'peak압력')
ax2.plot(cu, marker='o', markerfacecolor='green', markersize=10, color='green', linewidth=2, label='최소쿠션')
ax1.legend(loc='best')
ax2.legend(loc='best')
plt.show()
```

```python
plt.style.use('ggplot')

fig = plt.figure(figsize=(8,4))
#PEAK압력	최소쿠션	V-P위치	V-P압력	사출시간	계량시간	사이클타임	금형(상)	금형(하)
plt.boxplot(x=[df1['PEAK압력'],df1['최소쿠션'], df1['V-P위치'], df1['V-P압력'], df1['사출시간'],df1['계량시간'],df1['사이클타임'],df1['금형(상)'], df1['금형(하)']])
plt.title('box plot')
plt.ylabel('dd')
plt.show()
```
